Crime patterns have long been studied in criminology, with researchers investigating socio-economic, environmental, and psychological factors that influence criminal behavior. Among these, temperatures had been found to correlate with crime rate changes to some extent. Some studies suggest that warmer temperatures increase crime rates due to greater outdoor activity, increased social interactions, and heightened physiological responses to heat. Conversely, colder temperatures may suppress certain crimes, as people spend more time indoors. While previous research has explored seasonal crime trends at national or regional levels, there remains a gap in understanding localized effects across different cities and neighborhoods.

To address this gap, our project will examine crime patterns in Chicago, Los Angeles, and Austin, three cities with distinct climate conditions but comparable urban structures. By analyzing crime variations across different temperature patterns and neighborhoods, we aim to better understand how seasonal changes influence different types of crimes. To reflect the present reality, we choose to study the past 4 years of data.

### The Temperature-Crime Relationship
Numerous studies suggest the correlation between temperature and crime. Cohn & Rotton (2000)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) found that violent crimes tend to rise in summer, linking this to increased aggression and more social interactions outdoors. Similarly, Ranson (2014)<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) found that a 1°F increase in temperature was associated with a significant rise in violent crimes, particularly assaults and robberies, while property crimes showed weaker association.

However, the impact of temperature is not the same across all cities. Mares (2013)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) studied crime trends across U.S. cities and found that temperature-related crime fluctuations were more pronounced in regions with extreme seasonal variations compared to cities with milder climates. Additionally, a more recent study by Choi et al. (2024)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) examined temperature and crime correlations across multiple continents and found that cities with tropical/subtropical and humid climates, such as parts of Asia, had a stronger and more consistent positive correlation with crime rates. Meanwhile, in regions like the U.S. and Canada, where the climate varies and tends to be more temperate, some studies found a positive correlation while others reported no significant or even negative association.

### Comparing Cities & Crime Categories
Several studies have examined how different urban climates shape crime trends. Harries & Stadler (2017)<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) found that cold weather significantly reduced crime rates in New York, whereas the impact was smaller in Los Angeles due to its mild winters. These findings suggest that climate differences affect crime patterns differently across cities. Our research builds on this by analyzing crime variations in Chicago, Austin, and Los Angeles to understand how different climate conditions influence criminal activity.

Furthermore, past research has often overlooked the impact of urban structure and socio-economic factors on seasonal crime trends. Hipp et al. (2019)<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) and Block & Block (2018)<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) found that crime distribution varies based on land use, accessibility to public spaces, and population density. Our study builds on this by analyzing how crime patterns change across different Chicago neighborhoods, providing a localized perspective on seasonal crime fluctuations.

We already know that our dataset includes 34 crime categories, allowing us to analyze how temperature influences different types of offenses. We plan to categorize the 34 types of crimes into: Violent Crimes (e.g. Battery, Assault, Homicide, Criminal Sexual, etc.), Property Crimes (Theft, Motor Vehicle Theft, Burglary, etc.), Public Order & Regulatory Crimes (Weapons Violation, Liquor Law Violation, etc.), and Other Crimes & Offenses (Narcotics, Criminal Trespass, Deceptive Practice, Human Trafficking, etc.). By categorizing these crimes, we can better understand how different offenses respond to seasonal temperature changes.

### Contributions & Research Gaps
While prior studies have explored seasonal crime fluctuations, few have directly compared cities with different climate conditions or examined crime variations at the neighborhood level. Our research fills these gaps by:
- Comparing temperature-crime relationships in Chicago, Los Angeles, and Austin, three cities with distinct climate conditions but similar urban structures.
- Analyzing neighborhood-level crime variations within Chicago, examining how different areas are affected by seasonal crime trends.
- Exploring how socio-economic factors and city planning contribute to seasonal crime fluctuations, considering variations in population density, economic activity, and law enforcement presence.

By addressing these questions, our project aims to provide better understanding for criminology and urban planning, helping policymakers and law enforcement better understand seasonal crime trends and develop strategies tailored to different urban environments.
